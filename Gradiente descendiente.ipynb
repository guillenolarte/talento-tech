```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función de costo y su gradiente
def funcion_costo(x):
    return x**2

def gradiente(x):
    return 2*x
```

```python
# Gradiente descendente
def gradiente_descendente(learning_rate, iteraciones, x_inicial):
    x = x_inicial
    valores_x = [x]
    valores_costo = [funcion_costo(x)]

    for i in range(iteraciones):
        grad = gradiente(x)
        x = x - learning_rate * grad
        # Imprimir la iteración y el valor de x y f(x)
        print(f"Iteración {i+1}: x = {x}, f(x) = {funcion_costo(x)}")

        valores_x.append(x)
        valores_costo.append(funcion_costo(x))

    return valores_x, valores_costo
```

```python
# Parámetros
learning_rate = 0.2
iteraciones = 40
x_inicial = 10

# Ejecutar gradiente descendente y obtener valores para graficar
valores_x, valores_costo = gradiente_descendente(learning_rate, iteraciones, x_inicial)

# Graficar la función de costo y el proceso de gradiente descendente
x_vals = np.linspace(-10, 10, 400)
y_vals = funcion_costo(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label="Función de costo: $f(x) = x^2$")
plt.scatter(valores_x, valores_costo, color='red', label="Gradiente descendente", zorder=5)
plt.plot(valores_x, valores_costo, color='red', linestyle='--', zorder=4)
plt.title("Gradiente descendente sobre la función $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
