```python
import pygame
import random

# Inicializar pygame
pygame.init()

# Configurar los colores
BLANCO = (255, 255, 255)
ROJO = (255, 0, 0)
AZUL = (0, 0, 255)

# Tamaño de la pantalla
ANCHO_PANTALLA = 800
ALTO_PANTALLA = 600

# Crear la pantalla
pantalla = pygame.display.set_mode((ANCHO_PANTALLA, ALTO_PANTALLA))
pygame.display.set_caption("Atrapa al cuadrado")

# Configurar el reloj
reloj = pygame.time.Clock()

# Posiciones iniciales
jugador_x = ANCHO_PANTALLA // 2
jugador_y = ALTO_PANTALLA // 2
jugador_velocidad = 5

cuadrado_x = random.randint(0, ANCHO_PANTALLA - 30)
cuadrado_y = random.randint(0, ALTO_PANTALLA - 30)
cuadrado_tamaño = 30

puntuacion = 0

# Bucle principal del juego
ejecutando = True
while ejecutando:
    # Manejo de eventos
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False

    # Obtener las teclas presionadas
    teclas = pygame.key.get_pressed()

    # Movimiento del jugador
    if teclas[pygame.K_LEFT]:
        jugador_x -= jugador_velocidad
    if teclas[pygame.K_RIGHT]:
        jugador_x += jugador_velocidad
    if teclas[pygame.K_UP]:
        jugador_y -= jugador_velocidad
    if teclas[pygame.K_DOWN]:
        jugador_y += jugador_velocidad

    # Mantener al jugador dentro de la pantalla
    if jugador_x < 0:
        jugador_x = 0
    if jugador_x > ANCHO_PANTALLA - cuadrado_tamaño:
        jugador_x = ANCHO_PANTALLA - cuadrado_tamaño
    if jugador_y < 0:
        jugador_y = 0
    if jugador_y > ALTO_PANTALLA - cuadrado_tamaño:
        jugador_y = ALTO_PANTALLA - cuadrado_tamaño

    # Comprobar colisión con el cuadrado rojo
    if (jugador_x < cuadrado_x + cuadrado_tamaño and
        jugador_x + cuadrado_tamaño > cuadrado_x and
        jugador_y < cuadrado_y + cuadrado_tamaño and
        jugador_y + cuadrado_tamaño > cuadrado_y):
        cuadrado_x = random.randint(0, ANCHO_PANTALLA - cuadrado_tamaño)
        cuadrado_y = random.randint(0, ALTO_PANTALLA - cuadrado_tamaño)
        puntuacion += 1

    # Rellenar la pantalla
    pantalla.fill(BLANCO)

    # Dibujar el jugador (cuadrado azul)
    pygame.draw.rect(pantalla, AZUL, (jugador_x, jugador_y, cuadrado_tamaño, cuadrado_tamaño))

    # Dibujar el cuadrado rojo (objetivo)
    pygame.draw.rect(pantalla, ROJO, (cuadrado_x, cuadrado_y, cuadrado_tamaño, cuadrado_tamaño))

    # Mostrar la puntuación
    fuente = pygame.font.SysFont(None, 36)
    texto_puntuacion = fuente.render(f'Puntuación: {puntuacion}', True, (0, 0, 0))
    pantalla.blit(texto_puntuacion, (10, 10))

    # Actualizar la pantalla
    pygame.display.flip()

    # Controlar la velocidad de fotogramas
    reloj.tick(60)

# Cerrar el juego
pygame.quit()
```
